```python
import pandas as pd
import matplotlib.pyplot as plt #Importa libreria matplotlib para visualizacion de datos
import seaborn as sns
import plotly.express as px

#Carga los datos limpios
data_path = "../data/processed/cleaned_products.csv"
df = pd.read_csv(data_path) #Carga los datos del archivo en un DataFrame

#Analisis y visualizacion
print("Basic Stadistics")
print(df.describe()) #Mostrar resumen estadistico con el DataFrame

#Histograma de precios
plt.hist(df["price"], bins=20, edgecolor="black") #Crea un histograma de la columna "price" con 20 barras y bordes negros
plt.title("Distribucion de precios") #Añade titulo al histograma
plt.xlabel("Precio") #Añade etiqueta al "eje X"
plt.ylabel("Frecuencia") #Añade etiqueta al "eje Y"
plt.show() #Imprime el histograma

#Productos mas caros
print("Productos mas caros: ")
highestPrices = df.nlargest(6, "price")
print(highestPrices)


#Grafico de barras con Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x='title', y='price', data=df)
plt.xticks(rotation=45, ha='right')
plt.title('Precios de Productos')
plt.xlabel('Producto')
plt.ylabel('Precio ($)')
plt.show()

#Grafico de dispersión con Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='price', y='description', data=df)
plt.title('Precio vs Descripción')
plt.xlabel('Precio ($)')
plt.ylabel('Descripción')
plt.show()

#Gráfico de dispersión interactivo con Plotly
fig = px.scatter(df, x='price', y='description', hover_data=['title'])
fig.update_layout(title='Precio vs Descripción', xaxis_title='Precio ($)', yaxis_title='Descripción')
#fig.show()
fig.write_html('scatter_plot.html')
```
